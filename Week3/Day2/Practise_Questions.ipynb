# Task 3 finding unique values + count order

```python
customer_order=["Mango","Dates",
"Apple",
                "PineApple","Apple",
"Dates",
                 "Mango","Dates",
"Apple"]
```

```python
counts={}
#counting values
for item in customer_order:
    if item in counts:
        counts[item] += 1
    else:
        counts[item] = 1
# Unique Values
unique_items =  (counts.keys())
```

```python
print("Unique Items",unique_items)
print("Total Unique item :",len(unique_items))
print("Order Counts",counts)
```
