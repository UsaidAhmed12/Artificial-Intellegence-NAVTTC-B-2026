# Python Dictionary

# create a dictionary

```python
# creating a dictionary
country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}

# printing the dictionary
print(country_capitals)
```

# Access dictionary items

```python
# creating a dictionary
country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}

print(country_capitals["Germany"])
print(country_capitals["Canada"])
```

# Add items to dictionary

```python
# creating a dictionary
country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}

country_capitals["Italy"] = "Romee"
print(country_capitals)
```

# Remove dictionary items

```python
# creating a dictionary
country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}

del country_capitals["Germany"]
print(country_capitals)
```

# Clear method which will clean whole dictionary to empty

```python
# creating a dictionary
country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}

country_capitals.clear()
print(country_capitals)
```

# Change position of dictionary items or simply update key value to the last

```python
country_capitals = {
  "Germany": "Berlin", 
  "Italy": "Naples", 
  "England": "London"
}

country_capitals["Italy"] = "Berlin"
print(country_capitals)
```

# Iterate Through a dictionary using for loop 

```python
country_capitals = {
  "United States": "Washington D.C.", 
  "Italy": "Rome" 
}

# print dictionary keys one by one
for country in country_capitals:
    print(country)

print()

# print dictionary values one by one
for country in country_capitals:
    capital = country_capitals[country]
    print(capital)
```

# Dictionary length

```python
# Find dictionary length
country_capitals = {"England": "London", "Italy": "Rome"}

# get dictionary's length
print(len(country_capitals))     
```

# Dictionary Membership Test or simply condition check

```python
file_types = {
    ".txt": "Text File",
    ".pdf": "PDF Document",
    ".jpg": "JPEG Image",
}

# use of in and not in operators
print(".pdf" in file_types)      
print(".mp3" in file_types)       
print(".mp3" not in file_types)   
print(".mp4" in file_types)
```
