# Python List Comprehension

# list which doubled the number given in the list

```python
numbers = [1,2,3,4]
```

```python
double_numbers = [num * 2 for num in numbers]
```

```python
print(double_numbers)
```

# Syntax of List Comprehension

# [expression for item in list if condition == True]

 # USING  FOR LOOP

```python
number = [1,2,3,4,5]
```

```python
square_number = []
```

```python
for num in number:
    square_number.append(num * num)
```

```python
print(square_number)
```

# Using List Comprehension

```python
number = [1,2,3,4,5]
```

```python
square_number = [num * num for num in number]
```

```python
print(square_number)
```

# Conditionals in List Comprehension

```python
even_numbers = [num for num in range(1,10) if num % 2 == 0]
```

```python
print(even_numbers)
```

# List Comprehension with String

```python
word ="python"
```

```python
vowels = "aeiou"
```

```python
result =[char for char in word if char in vowels]
```

```python
print(result)
```

# Nested List Comprehension

```python
multiplication = [[i*j for j in range(1,6)] for i in range(2,5)]
```

```python
print(multiplication)
```

# Equivalent Nested for Loop

```python
multiplication = []

for i in range(2, 5):
    row = []
    for j in range(1, 6):
        row.append(i * j)
    multiplication.append(row)

print(multiplication)
```
