# Number in list is even or odd

```python
a=[1,2,3,4,6,8,9,0]
print(a)
```

```python
print("Even number is")
for num in a:
    if num % 2 == 0:
        print(num)
print("odd number is:")
for num in a:
    if num % 2 != 0:
        print(num)
```

```python
print("Even number is")
for num in a:
    if num % 2 == 0:
        print(num)
```

```python
print("Even and odd number is")
for num in a:
    if num % 2 == 0:
        print(num,"is even")
    else:
        print(num,"is odd")
```

```python
minimum = a[0]
maximum = a[0]
second_min = None
```

```python
# using for loop
for i in range (len(a)):
    if a[i] < minimum:
        minimum = a[i]
    elif a[i] > maximum:
        maximum = a[i]

print("Maximum value is", maximum)
print("Minimum value is", minimum)

for i in range(len(a)):
    # second minimum
    if a[i] != minimum:
        if second_min is None or a[i] < second_min:
            second_min = a[i]

print("2nd Minimum value is", second_min)
            
    
    
```
